```python
import pandas as pd
datos = pd.read_csv('altura.csv')
datos.columns = datos.columns.str.strip()
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(x="altura", y="peso", data=datos, hue="peso", palette="coolwarm")
```

```python
#Caracteristicas (X), etiqueta (y)
X = datos["altura"]
y = datos["peso"]
```

```python
y
```

```python
X
```

```python
#X,values
X_procesada = X.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(X_procesada, y_procesada)
```

```python
altura = 195
prediccion = modelo.predict([[altura]])
print(f"la altura es {altura} cm y el peso posible{prediccion} kg")
```

```python
modelo.score(X_procesada, y_procesada)
```
